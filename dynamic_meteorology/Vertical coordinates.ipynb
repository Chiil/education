# Changing vertical coordinates
In this tutorial...

We take the potential temperature $\theta \left( x, z \right)$ for this example.

Casting $\theta \left( x, z \right)$ into $\theta \left( x, p \right)$

___
1. Under which conditions can we rewrite $\theta \left( x, z \right)$ into $\theta \left( x, p \right)$?
___

We can now say that $z \left( x, p\right)$, thus $p \left( x, z \right) = p \left( x, z \left(x, p \right) \right)$

$$\left( \dfrac{\partial p}{\partial x} \right)_z = 
  \left( \dfrac{\partial p}{\partial x} \right)_p + \dfrac{\partial p}{\partial z} \dfrac{\partial z}{\partial x}$$

```python
import numpy as np
import matplotlib.pyplot as pl
%pylab inline
pl.rcParams.update({'font.size': 12})          # Set the standard font size of the plots to 11pt.
pl.rcParams.update({'figure.figsize': [13,6]}) # Set the standard figure size.
```

```python
x = np.linspace(0., 11000., 100)
z = np.linspace(0., 11000., 100)
xx, zz = np.meshgrid(x, z)
theta = 300. + 0.006*zz + 2.*(1.-zz/11000.)*np.sin(2.*np.pi/5500.*xx)
p = 1.e5 * np.exp(-zz/5000.)
```

```python
pl.contourf(xx, zz, p)
pl.colorbar()
pl.contour(xx, zz, p, colors='k')
```

```python
pl.subplot(121)
pl.contourf(xx, zz, theta)
pl.colorbar()
pl.contour(xx, zz, theta, colors='k')
pl.xlabel('x')
pl.ylabel('z')
pl.title('potential temperature in height coordinates');
pl.subplot(122)
pl.contourf(xx, p, theta)
pl.colorbar();
pl.contour(xx, p, theta, colors='k')
pl.gca().invert_yaxis()
#pl.gca().set_yscale('log')
pl.xlabel('x')
pl.ylabel('p')
pl.title('potential temperature in pressure coordinates');
```
